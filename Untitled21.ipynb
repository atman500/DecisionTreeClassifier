```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import google.colab
```

```python
from google.colab import files
from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
from sklearn.tree import export_graphviz
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('‏‏canceR_uae .csv',sep=';')
```

```python
print(data.head(5))
```

```python
data=data.drop('Comorbidities',axis=1)
```

```python
print(data.head())
```

```python
dep=data.iloc[:,7]
```

```python
print(dep.head(5))
```

```python
indep=data.drop('Outcome',axis=1)
```

```python
print(indep.head())
```

```python
data.describe()
```

```python
indep=pd.get_dummies(indep)
```

```python
print(indep.head())
```

```python
model1=DecisionTreeClassifier()
model1.fit(indep,dep)
```

```python
dep_pred=model1.predict(indep)
```

```python
print(model1.get_depth())
print(model1.get_n_leaves())
print(model1.feature_names_in_)
print(model1.feature_importances_)
```

```python
feature_importances=pd.DataFrame(model1.feature_importances_,index=indep.columns,columns=['importance'])
print(feature_importances)
```

```python
print(confusion_matrix(dep,dep_pred))
```

```python
print(classification_report(dep,dep_pred))
```

Coorectionof the model1

```python
model2=DecisionTreeClassifier(
    max_depth=15,
    min_samples_leaf=12,
    min_samples_split=12
)
model2.fit(indep,dep)
```

```python
print(model2.get_depth())
print(model2.get_n_leaves())
print(model2.get_params())
```

```python
print(model2.get_depth())
print(model2.get_n_leaves())
print(model2.get_params())
print(model2.feature_names_in_)
print(model2.feature_importances_)
```

```python
feature_importances=pd.DataFrame(model2.feature_importances_,index=indep.columns,columns=['importance'])
print(feature_importances)
```

```python
dep_predict_corrected=model2.predict(indep)
print(dep_predict_corrected)
print(accuracy_score(dep,dep_predict_corrected))
```

```python
print(classification_report(dep,dep_predict_corrected))
```

```python
plt.figure(figsize=(50,50))
plot_tree(model2,filled=True,rounded=True,feature_names=indep.columns)
plt.title('DEcisiontree model ')
plt.show()
```
