```python
model_3=DecisionTree=DecisionTreeClassifier(
    max_depth=18,
    min_samples_leaf=7,
    min_samples_split=7)
model_3.fit(indep,dep)
```

```python
dep_predict_3=model_3.predict(indep)
```

```python
print("Criteria of the Pruned Model: model_3")
print("=="*3)
print(confusion_matrix(dep,dep_predict_3))
print("\n")
print(classification_report(dep,dep_predict_3))
print("\n")
print('The accuracy value :',accuracy_score(dep,dep_predict_3))
```

```python
feature_names=model_3.feature_names_in_
feature_importances=model_3.feature_importances_
```

```python
features=pd.DataFrame({'feature_names':feature_names,'feature_importance':feature_importances})
```

```python
print("feature names Vs  feature importances ")
print("=="*3)
print(features.head(18))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sort features by importance
features = features.sort_values(by='feature_importance', ascending=False)

# Create the plot
plt.figure(figsize=(12, 10))
sns.barplot(x='feature_importance', y='feature_names', data=features)
plt.title('Feature Importances of Pruned Model (model_3)')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
print("Decison Tree Classifier of the Prunded model_3")
print("=="*3)
plt.figure(figsize=(20,20))
plot_tree(model_3,filled=False,feature_names=indep.columns,rounded=True,fontsize=15)
plt.title('Decision Tree Classifier with Pruning')
plt.show()
```

Usunig sklear.model selectio as  train and test

```python
from sklearn.model_selection import train_test_split
```

```python
indep_test,indep_train,dep_test,dep_train=train_test_split(indep,dep,test_size=0.2,random_state=0)
```

```python
model_trained=DecisionTreeClassifier(
    max_depth=15,
    min_samples_leaf=10,
    min_samples_split=10,
   )
model_trained.fit(indep_train,dep_train)
```

```python
dep_predict_trained=model_trained.predict(indep_test)
```

```python
print("Evaluation of the model_trained")
print("==="*3)
print(confusion_matrix(dep_test,dep_predict_trained))
print("\n")
print(classification_report(dep_test,dep_predict_trained))
print("\n")
print(accuracy_score(dep_test,dep_predict_trained))
```
